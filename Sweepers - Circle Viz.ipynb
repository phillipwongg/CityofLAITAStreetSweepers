```python
import pandas as pd
import sys
import os

from mapboxgl.viz import *
from mapboxgl.utils import *
```

```python
token = os.getenv('MAPBOX_ACCESS_TOKEN')
```

```python
df = pd.read_csv('sweeper_coordinate_test.csv',nrows=20)
df.dropna()
```

```python
# Create geojson data object
df_to_geojson(df,
              filename ="../DSF Street Sweeping Project/streetsweep_citations_days.geojson", 
              properties = ['Weekday','Issue Time', 'Ticket Number', 'Route'], 
              lat = 'Latitude', lon = 'Longitude',
              precision = 4)
```

```python
from mapboxgl.viz import *
from mapboxgl.utils import *

# Assign color stops
measure = 'Issue Time'
color_breaks = [round(df[measure].quantile(q=x*0.1), 2) for x in range(1,9)]
color_stops = create_color_stops(color_breaks, colors='YlOrRd')


# Initialize CircleViz with Categorical Measure Data
citations_viz = CircleViz('../DSF Street Sweeping Project/streetsweep_citations_days.geojson', 
                access_token=token, height='500px',
                radius = 2.5,
                color_property = "Issue Time",
                color_stops = color_stops,
                stroke_color = 'black',
                stroke_width = 0.2,
                center = (-118.2,34),
                zoom = 9,
                below_layer = 'waterway-label')
citations_viz.label_property = "Weekday"
citations_viz.stroke_width = 0
citations_viz.label_size = 8
citations_viz.style='mapbox://styles/mapbox/dark-v9?optimize=true'
citations_viz.label_color = 'hsl(0, 0%, 70%)'
citations_viz.label_halo_color = 'hsla(0, 0%, 10%, 0.75)'
citations_viz.show()
```
